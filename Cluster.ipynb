```python
import numpy as np
from sklearn.cluster import KMeans

X = np.array([[4,21],[5,19],[10,24],[4,17],[3,16],
              [11,25],[14,24],[6,22],[10,21],[12,21]])

kmeans = KMeans(n_clusters=2, init=np.array([[7,20],[8,25]]), n_init=1, max_iter=2)
kmeans.fit(X)

print("Final centroids:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)
```
