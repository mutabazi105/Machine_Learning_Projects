```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_excel("Real estate valuation data set.xlsx")

# Overview
print(df.shape)        # how many rows and columns
print(df.columns)      # column names
print(df.head())       # first 5 rows
print(df.describe())   # basic statistics
```

```python
X = df[["X3 distance to the nearest MRT station"]]  # input (must be 2D)
y = df["Y house price of unit area"]                 # target

model = LinearRegression()
model.fit(X, y)

print("Slope (m):    ", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
print("R²:", round(r2, 4))
```

```python
features = ["X1 transaction date", "X2 house age",
            "X3 distance to the nearest MRT station",
            "X4 number of convenience stores",
            "X5 latitude", "X6 longitude"]

for feat in features:
    X_f = df[[feat]]
    m = LinearRegression().fit(X_f, y)
    r2 = r2_score(y, m.predict(X_f))
    print(f"{feat}: R² = {r2:.4f}")
```
