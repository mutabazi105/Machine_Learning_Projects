```python
# English version of the code for Naive Bayes-----
# STEP 1: Import libraries


import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB


# Load the Iris dataset (Task 1)


df = pd.read_csv('Iris.csv')

print("\nTask 1:First 5 rows of the dataset:")
print(df.head())


# STEP 3: Separate features and target (Task 2)


# Features (X) - all columns except the last one (class/species)
X = df.iloc[:, :-1]  # or: X = df.drop('class', axis=1)

# Target (y) - the last column (class/species)
y = df.iloc[:, -1]   # or: y = df['class']

print(f"\nTask 2: Features (X) shape: {X.shape}")
print(f"   Target (y) shape: {y.shape}")
print(f"   Feature names: {X.columns.tolist()}")
print(f"   Unique classes: {y.unique()}")


# STEP 4: Train Gaussian Naive Bayes model (Task 3)

model = GaussianNB()
model.fit(X, y)

print("\nTask 3: Model trained successfully!")
print(f" Model class: {model.__class__.__name__}")


# STEP 5: Predict classes for rows 0, 50, 100 (Task 4)

rows_to_predict = [0, 50, 100]

print("\nTask 4: Predictions for specific rows:")
for row_idx in rows_to_predict:
    single_row = X.iloc[row_idx:row_idx+1]  # Keep as DataFrame
    prediction = model.predict(single_row)[0]
    print(f"   Row {row_idx}: Actual = {y.iloc[row_idx]}, Predicted = {prediction}")


# STEP 6: Show probabilities for those rows (Task 5)


print("\nTask 5: Probabilities for specific rows:")
for row_idx in rows_to_predict:
    single_row = X.iloc[row_idx:row_idx+1]
    probabilities = model.predict_proba(single_row)[0]
    print(f"\n   Row {row_idx}:")
    print(f"      Actual class: {y.iloc[row_idx]}")
    for class_name, prob in zip(model.classes_, probabilities):
        print(f"      P({class_name}) = {prob:.4f} ({prob*100:.2f}%)")


# STEP 7: Find the most uncertain prediction (Task 6)


# Get probabilities for ALL rows
all_probabilities = model.predict_proba(X)

# For each row, find the highest probability (the model's confidence)
max_probabilities = np.max(all_probabilities, axis=1)

# Find the row with the smallest max probability (most uncertain)
most_uncertain_idx = np.argmin(max_probabilities)
most_uncertain_prob = max_probabilities[most_uncertain_idx]
most_uncertain_prediction = model.predict(X.iloc[most_uncertain_idx:most_uncertain_idx+1])[0]

print("\nTask 6: Most uncertain prediction:")
print(f"   Row index: {most_uncertain_idx}")
print(f"   Highest class probability: {most_uncertain_prob:.4f} ({most_uncertain_prob*100:.2f}%)")
print(f"   Predicted class: {most_uncertain_prediction}")
print(f"   Actual class: {y.iloc[most_uncertain_idx]}")
print(f"   All probabilities for this row:")
for class_name, prob in zip(model.classes_, all_probabilities[most_uncertain_idx]):
    print(f"P({class_name}) = {prob:.4f} ({prob*100:.2f}%)")


# BONUS: Summary statistics


print("\n" + "="*50)
print(" SUMMARY")
print("="*50)
print(f"Total samples: {len(df)}")
print(f"Number of features: {X.shape[1]}")
print(f"Number of classes: {len(model.classes_)}")
print(f"Classes: {model.classes_}")
print(f"\nConfidence statistics:")
print(f"   Average confidence: {np.mean(max_probabilities):.4f}")
print(f"   Minimum confidence: {np.min(max_probabilities):.4f}")
print(f"   Maximum confidence: {np.max(max_probabilities):.4f}")

```

```python
# Deutsch version 

import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB

# Aufgabe 1: Daten laden und anzeigen
df = pd.read_csv('Iris.csv')
print("Aufgabe 1:")
print(df.head())

# Aufgabe 2: X und y trennen
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print(f"\nAufgabe 2: X Form={X.shape}, y Form={y.shape}")

# Aufgabe 3: Modell trainieren
model = GaussianNB()
model.fit(X, y)
print("\nAufgabe 3: Modell trainiert")

# Aufgabe 4: Vorhersagen für Zeilen 0,50,100
print("\nAufgabe 4:")
for zeile in [0,50,100]:
    vorhersage = model.predict(X.iloc[zeile:zeile+1])[0]
    print(f"Zeile {zeile}: {vorhersage} (tatsächlich: {y.iloc[zeile]})")

# Aufgabe 5: Wahrscheinlichkeiten
print("\nAufgabe 5:")
for zeile in [0,50,100]:
    prob = model.predict_proba(X.iloc[zeile:zeile+1])[0]
    print(f"Zeile {zeile}: {dict(zip(model.classes_, prob))}")

# Aufgabe 6: Unsicherste Vorhersage
print("\nAufgabe 6:")
alle_probs = model.predict_proba(X)
max_probs = np.max(alle_probs, axis=1)
unsicherste = np.argmin(max_probs)
print(f"Unsicherste Zeile: {unsicherste} (Sicherheit: {max_probs[unsicherste]:.4f})")
```
