## Schritt 6: Features und Zielvariable trennen

**Erklärung:**  
Jetzt trennen wir die Eingabedaten von der Zielvariable. `X` enthält die Merkmale der Blume, also zum Beispiel Länge und Breite von Kelch- und Blütenblättern. `y` enthält die Klasse, also die Iris-Art.

````python
# Alternative 1: Using .drop()
X = df.drop(columns='Species')
y = df['Species']

# Alternative 2: Using .iloc (by position)
X = df.iloc[:, :-1]  # All columns except the last one
y = df.iloc[:, -1]   # Only the last column
````

## Schritt 7: Daten in Trainings- und Testdaten aufteilen

**Erklärung:**  
Die Daten werden in Trainingsdaten und Testdaten aufgeteilt. Mit den Trainingsdaten lernt das Modell. Mit den Testdaten prüfen wir später, wie gut das Modell neue, unbekannte Daten klassifizieren kann. Hier verwenden wir 80 Prozent Training und 20 Prozent Test.



````python
from sklearn.model_selection import train_test_split

# Split data into training and test sets
# X = input data (features)
# y = target values (classes)

# We split the data into two parts:
# - Training data → for the model to learn from
# - Test data → to check how good the model is

X_train, X_test, y_train, y_test = train_test_split(
    
    X,  # Input data (e.g., SepalLength, PetalWidth, ...)
    
    y,  # Target variable (e.g., Iris species)
    
    test_size=0.2,  # 20% of data will be used for testing
                    # 80% remains for training
    
    random_state=42 # Ensures the split is always the same (reproducible)
)
````

````python
print("Data Split Summary:")
print(f"Training data - X_train: {X_train.shape}")
print(f"Training data - y_train: {y_train.shape}")
print(f"Test data - X_test: {X_test.shape}")
print(f"Test data - y_test: {y_test.shape}")
print()
print(f"Training samples: {len(X_train)} ({len(X_train)/len(X)*100:.0f}%)")
print(f"Test samples: {len(X_test)} ({len(X_test)/len(X)*100:.0f}%)")
````

````python
# Daten in Trainings- und Testdaten aufteilen

# X = Eingabedaten (Features)
# y = Zielwerte (Klassen)

# Wir teilen die Daten in zwei Teile:
# - Trainingsdaten → zum Lernen des Modells
# - Testdaten → zum Überprüfen, wie gut das Modell ist

X_train, X_test, y_train, y_test = train_test_split(
    
    X,  # Eingabedaten (z. B. SepalLength, PetalWidth, ...)
    
    y,  # Zielvariable (z. B. Iris-Art)
    
    test_size=0.2,  # 20% der Daten werden für den Test verwendet
                    # 80% bleiben fürs Training
    
    random_state=42 # sorgt dafür, dass die Aufteilung immer gleich ist
)
````

## Schritt 8: Decision Tree Modell erstellen

**Erklärung:**  
Hier erstellen wir das Modell. Ein Decision Tree trifft Entscheidungen anhand von Regeln. Mit `max_depth=3` begrenzen wir die Tiefe des Baums. Dadurch wird das Modell einfacher und Overfitting kann reduziert werden.

**Aufgabe:**  
### Aufgabe

Erstelle einen Decision Tree Classifier mit folgenden Parametern:

- `max_depth=3` (maximale Tiefe des Baums)
- `random_state=42` (für reproduzierbare Ergebnisse)
- `min_samples_leaf=10` (mindestens 10 Datenpunkte pro Blatt)

Speicheren Sie das Modell in einer Variable namens `model`.


**Lösung:**


````python
from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree model with specified parameters
model = DecisionTreeClassifier(
    max_depth=3,           # Maximum depth of the tree (limits complexity)
    random_state=42,       # Ensures reproducible results
    min_samples_leaf=10    # Minimum number of samples per leaf node
)
````

````python
# Display the model parameters
print("Decision Tree Model created with parameters:")
print(f"   max_depth: {model.max_depth}")
print(f"   random_state: {model.random_state}")
print(f"   min_samples_leaf: {model.min_samples_leaf}")
print(f"\n Model ready for training!")
````

## Schritt 12: Decision Tree visualisieren

**Erklärung:**
